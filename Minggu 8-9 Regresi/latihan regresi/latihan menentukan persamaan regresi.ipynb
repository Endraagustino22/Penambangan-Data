```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Data
usia_mobil = [5, 4, 6, 5, 5, 5, 6, 6, 2, 7, 7]
harga_mobil = [85, 103, 70, 82, 89, 98, 66, 95, 169, 70, 48]
```

```python
# Membuat DataFrame
df = pd.DataFrame({
    'Usia Mobil (tahun)': usia_mobil,
    'Harga Mobil ($100)': harga_mobil
})

# Menampilkan DataFrame
print(df)
```

```python
X = df[['Usia Mobil (tahun)']]  # Fitur (usia mobil)
y = df['Harga Mobil ($100)']    # Target (harga mobil)

# Membuat model regresi linier
model = LinearRegression()

# Melatih model
model.fit(X, y)

# Mendapatkan koefisien (m) dan intersep (b)
m = model.coef_[0]
b = model.intercept_

# Menampilkan persamaan regresi
print(f"Persamaan regresi: y = {m:.2f} * x + {b:.2f}")
```

```python

# Prediksi harga mobil berdasarkan regresi
y_pred = model.predict(X)

# Plot data asli
plt.scatter(X, y, color='blue', label='Data Asli')

# Plot garis regresi
plt.plot(X, y_pred, color='red', label='Garis Regresi')

# Menambahkan label dan judul
plt.xlabel('Usia Mobil (tahun)')
plt.ylabel('Harga Mobil ($100)')
plt.title('Regresi Linier Usia Mobil vs Harga Mobil')
plt.legend()

# Menampilkan plot
plt.show()

```
